```python
#Exploratory Data Analysis (EDA):
#Top 7 RICE PRODUCTION State Data(Bar_plot)
data_rice = df[["State Name","Year","RICE PRODUCTION (1000 tons)"]]
```

```python
# Rice Production By West Bengal Districts
data_westbengal = df[(df["State Name"]=="West Bengal")][["Dist Name","Year","RICE PRODUCTION (1000 tons)"]]
westbengal_rice = data_westbengal.groupby("Dist Name")["RICE PRODUCTION (1000 tons)"].sum().reset_index()
westbengal_rice = westbengal_rice.sort_values("RICE PRODUCTION (1000 tons)",ascending=False)
```

```python
# Top 10 Wheat Production Years From UP
data_uttarpradesh = df[df["State Name"] == "Uttar Pradesh"][["Year", "WHEAT PRODUCTION (1000 tons)"]]
uttarpradesh_wheat = data_uttarpradesh.groupby("Year")["WHEAT PRODUCTION (1000 tons)"].sum().reset_index()
top10_wheat = uttarpradesh_wheat.sort_values("WHEAT PRODUCTION (1000 tons)", ascending=False).head(10)
```

```python
# Soyabean Production by Top 5 States and Yield Efficiency
data_soyabean = df.groupby("State Name")[[
    "SOYABEAN PRODUCTION (1000 tons)",
    "SOYABEAN AREA (1000 ha)"
]].sum().reset_index()

# Calculate yield efficiency
data_soyabean["Yield Efficiency (tons/ha)"] = data_soyabean["SOYABEAN PRODUCTION (1000 tons)"] 
data_soyabean["SOYABEAN AREA (1000 ha)"]
top5_soyabean = data_soyabean.sort_values("SOYABEAN PRODUCTION (1000 tons)", ascending=False).head(5)
# Line Chart for Soyabean Production
# Line Plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=top5_soyabean, x="SOYABEAN PRODUCTION (1000 tons)", y ="State Name" , color= "lightgreen", marker = "o")
plt.title("Top 5 Soyabean Producing States in India",fontsize=14)
plt.xlabel("Total Soyabean Production(1000 tons)")
plt.ylabel("State Name")
plt.legend()
plt.show()
```

```python
# Bar Chart for Yield Efficiency
x = np.arange(len(top5_soyabean["State Name"]))
width = 0.35  
plt.figure(figsize=(10,6))
plt.bar(x + width/2, top5_soyabean["Yield Efficiency (tons/ha)"], width, label="Yield Efficiency", color="red")
plt.xticks(x, top5_soyabean["State Name"], rotation=30, ha="right")
plt.ylabel("Value")
plt.title("Soyabean Yield Efficiency (Top 5 States)")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Impact of Area Cultivated on Production (Rice, Wheat, Maize)
data_area_production = df.groupby("Year")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)",
    "MAIZE PRODUCTION (1000 tons)", "MAIZE AREA (1000 ha)"
]].sum().reset_index()
```

```python
# Regression Line Plot
plt.figure(figsize=(12,6))
sns.regplot(x="RICE AREA (1000 ha)", y="RICE PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Rice", color="blue")
sns.regplot(x="WHEAT AREA (1000 ha)", y="WHEAT PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Wheat", color="red")
sns.regplot(x="MAIZE AREA (1000 ha)", y="MAIZE PRODUCTION (1000 tons)", data=data_area_production, scatter_kws={'s':40}, label="Maize", color="yellow")

plt.xlabel("Area Cultivated (1000 ha)")
plt.ylabel("Production (1000 tons)")
plt.title("Correlation between Area and Production")
plt.legend()
plt.show()
```

```python
# Rice vs. Wheat Yield Across States 
data_yield = df.groupby("State Name")[[
    "RICE PRODUCTION (1000 tons)", "RICE AREA (1000 ha)",
    "WHEAT PRODUCTION (1000 tons)", "WHEAT AREA (1000 ha)"
]].sum().reset_index()

data_yield["Rice Yield"] = data_yield["RICE PRODUCTION (1000 tons)"] / data_yield["RICE AREA (1000 ha)"]
data_yield["Wheat Yield"] = data_yield["WHEAT PRODUCTION (1000 tons)"] / data_yield["WHEAT AREA (1000 ha)"]

data_yield = data_yield[["State Name", "Rice Yield", "Wheat Yield"]]
```

```python
# Heatmap Plot
data_heat = data_yield.set_index("State Name")[["Rice Yield","Wheat Yield"]]

plt.figure(figsize=(8,10))
sns.heatmap(data_heat, annot=True, cmap="tab10", fmt=".2f")
plt.title("Rice vs Wheat Yield Across States (Heatmap)")
plt.show()
```
